```python
import pandas as pd
```

```python
#reading excel file
```

```python
df = pd.read_excel(r"C:\Users\shrut\OneDrive\Documents\Pandas project\Data Cleaning using Pandas project\Customer Call List (1).xlsx")
```

```python
#dropping duplicates
```

```python
df = df.drop_duplicates()
```

```python
#dropping unwanted columns
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#removing unwanted characters from Last Name column using strip function
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

```python
#removing unwanted values in phone number
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
#applying lambda function to split phone number and converting the phone number column to string
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
#splitting the phone number in accepted format
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#replacing unwanted values with empty characters
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
df['Address'].str.split(',', n=1, expand=True)
```

```python
#splitting address into street address, state and zip
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2,  expand=True)
df
```

```python
#replacing Yes with Y
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df
```

```python
#replacing No with N
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
#replacing N/a with empty spaces
```

```python
df = df.replace('N/a', '')
```

```python
df = df.fillna('')
```

```python
df
```

```python
#the requirement is to get the list of customers whom the call center can call, hence dropping the ones which says Y in Do_Not_contact column
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)
```

```python
#same for phonenumber
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)
```

```python
#resetting index
```

```python
df = df.reset_index(drop=True)
df
```
